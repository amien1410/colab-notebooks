```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA, TruncatedSVD
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features: sepal length, sepal width, petal length, petal width
y = iris.target  # Target: species

# Convert to DataFrame for readability
df = pd.DataFrame(X, columns=iris.feature_names)
print("Original Data Shape:", df.shape)
```

```python
# Select top 2 features using ANOVA F-test
selector = SelectKBest(score_func=f_classif, k=2)
X_selected = selector.fit_transform(X, y)

print("Selected Features Shape:", X_selected.shape)
```

```python
# Reduce dimensions to 2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("PCA Transformed Shape:", X_pca.shape)

# Visualize
plt.figure(figsize=(6, 4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title("PCA: Principal Component View")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```

```python
# Truncated SVD (useful for sparse data or text)
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X)

print("SVD Transformed Shape:", X_svd.shape)

# Visualize
plt.figure(figsize=(6, 4))
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, cmap='plasma', edgecolor='k')
plt.title("Truncated SVD: Compressed View")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)
plt.show()
```
