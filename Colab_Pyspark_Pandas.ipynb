```python
import pandas as pd
data = {'Country': ['United States', 'Brazil', 'Russia'], 'River':
['Mississippi', 'Amazon', 'Volga']}
pandas_df = pd.DataFrame(data)

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("BigRivers").getOrCreate()
pyspark_df = spark.createDataFrame(pandas_df)
pyspark_df.show()
```

```python
# Pandas selecting columns
import pandas as pd
pandas_df = pd.DataFrame({'President': ['George Washington', 'AbrahamLincoln', 'Franklin D. Roosevelt', 'John F. Kennedy', 'Barack Obama'], 'Year of Office': [1789, 1861, 1933, 1961, 2009]})
pandas_df[["President", "Year of Office"]]
```

```python
# Pyspark selecting columns
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Presidents").getOrCreate()
spark_df = spark.createDataFrame([
("George Washington", 1789),
("Abraham Lincoln", 1861),
("Franklin D. Roosevelt", 1933),
("John F. Kennedy", 1961),
("Barack Obama", 2009)],
["President", "Year of Office"])
spark_df[["President", "Year of Office"]].show()
```

```python
# Pyspark select specific columns
spark_df.select("President", "Year of Office").show()
```

```python
# Pandas select specific columns
pandas_df.filter(items = ["President", "Year of Office"])
```

```python
# Aggregating data with Pandas
import pandas as pd
pandas_data = {
'movie': ['Avengers', 'Frozen', 'Star Wars',
'The Lion King', 'Harry Potter'],
'revenue': [90000000, 70000000, 110000000, 80000000, 100000000]}
pandas_df = pd.DataFrame(pandas_data)
agg_pandas_df = pandas_df.groupby('movie').agg({'revenue':'mean'})
agg_pandas_df
```

```python
# Aggregating data with Pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("MovieRevenueAggregation").getOrCreate()
spark_data = [('Avengers', 90000000), ('Frozen', 70000000), ('Harry Potter', 100000000), ('Star Wars', 110000000), ('The Lion King', 80000000)]
spark_df = spark.createDataFrame(spark_data, ['movie', 'revenue'])
agg_spark_df = spark_df.groupBy('movie').agg({'revenue': 'mean'})
agg_spark_df.show()
```
